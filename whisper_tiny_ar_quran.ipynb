## **ASR**

```python
import torch
from transformers import pipeline

def transcribe_egyptian_audio(audio_path):
    device = "cuda:0" if torch.cuda.is_available() else "cpu"

    model_id = "MAdel121/whisper-small-egyptian-arabic"

    print(f"Loading model on {device}...")
    pipe = pipeline(
        "automatic-speech-recognition",
        model=model_id,
        device=device
    )

    print("Transcribing...")
    result = pipe(
        audio_path,
        chunk_length_s=30,  # لتقسيم الملفات الطويلة
        batch_size=8,        # لتسريع العملية
        generate_kwargs={"language": "arabic"}
    )

    return result["text"]

file_path = "test.ogg"
try:
    text = transcribe_egyptian_audio(file_path)
    print("-" * 30)
    print("النص المستخرج:")
    print(text)
    print("-" * 30)
except Exception as e:
    print(f"حصلت مشكلة: {e}")
```
